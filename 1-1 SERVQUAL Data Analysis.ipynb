```r
library(xlsx)
library(ggplot2)
```

```r
df_servqual <- read.xlsx('df_servqual.xlsx', sheetName = 1)
```

```r
df_servqual
```

```r
str(df_servqual)
```

```r
df_servqual$ID <- as.factor(df_servqual$ID)
```

```r
avg_colokan_real <- mean(df_servqual$colokan_real)
avg_colokan_hope <- mean(df_servqual$colokan_hope)
avg_wifi_real <- mean(df_servqual$wifi_real)
avg_wifi_hope <- mean(df_servqual$wifi_hope)
avg_food_real <- mean(df_servqual$food_real)
avg_food_hope <- mean(df_servqual$food_hope)
avg_price_real <- mean(df_servqual$price_real)
avg_price_hope <- mean(df_servqual$price_hope)
avg_foodspeed_real <- mean(df_servqual$foodspeed_real)
avg_foodspeed_hope <- mean(df_servqual$foodspeed_hope)
avg_responsespeed_real <- mean(df_servqual$responsespeed_real)
avg_responsespeed_hope <- mean(df_servqual$responsespeed_hope)
avg_cashierspeed_real <- mean(df_servqual$cashierspeed_real)
avg_cashierspeed_hope <- mean(df_servqual$cashierspeed_hope)
avg_menu_real <- mean(df_servqual$menu_real)
avg_menu_hope <- mean(df_servqual$menu_hope)
avg_courtesy_real <- mean(df_servqual$courtesy_real)
avg_courtesy_hope <- mean(df_servqual$courtesy_hope)
avg_diningarea_real <- mean(df_servqual$diningarea_real)
avg_diningarea_hope <- mean(df_servqual$diningarea_hope)
avg_staffuniform_real <- mean(df_servqual$staffuniform_real)
avg_staffuniform_hope <- mean(df_servqual$staffuniform_hope)
avg_opentime_real <- mean(df_servqual$opentime_real)
avg_opentime_hope <- mean(df_servqual$opentime_hope)
avg_specialneeds_real <- mean(df_servqual$specialneeds_real)
avg_specialneeds_hope <- mean(df_servqual$specialneeds_hope)
```

```r
delta <- list()
delta[1] <- -(avg_colokan_real - avg_colokan_hope)
delta[2] <- -(avg_wifi_real - avg_wifi_hope)
delta[3] <- -(avg_food_real - avg_food_hope)
delta[4] <- -(avg_price_real - avg_price_hope)
delta[5] <- -(avg_foodspeed_real - avg_foodspeed_hope)
delta[6] <- -(avg_responsespeed_real - avg_foodspeed_hope)
delta[7] <- -(avg_cashierspeed_real - avg_cashierspeed_hope)
delta[8] <- -(avg_menu_real - avg_menu_hope)
delta[9] <- -(avg_courtesy_real - avg_courtesy_hope)
delta[10] <- -(avg_diningarea_real - avg_diningarea_hope)
delta[11] <- -(avg_staffuniform_real - avg_staffuniform_hope)
delta[12] <- -(avg_opentime_real - avg_opentime_hope)
delta[13] <- -(avg_specialneeds_real - avg_specialneeds_hope)

names <- list('plug', 'wifi', 'food', 'price', 'sfood', 'sresp', 'scash', 'menu', 'cour', 'dinar', 'staff', 'open', 'spec')
dimension <- 1
dimension[1:4] <- 'Red'
dimension[5:7] <- 'Orange'
dimension[8:9] <- 'Yellow'
dimension[10:11] <- 'Green'
dimension[12:13] <- 'Blue'
```

```r
df_results <- data.frame(row.names = names)
df_results$names <- names
df_results$delta <- delta
df_results$dimension <- dimension
```

```r
df_results_ordered <- df_results[order(as.numeric(df_results$delta), decreasing = TRUE), ]
```

```r
dim_names <- list('Reliability', 'Responsiveness', 'Assurance', 'Tangibles', 'Empathy')
df_dimResults <- data.frame(row.names = dim_names)
df_dimResults$dim_names <- dim_names
```

```r
dimResults <- 1
dimResults[1] <- mean(as.numeric(df_results$delta[1:4]))
dimResults[2] <- mean(as.numeric(df_results$delta[5:7]))
dimResults[3] <- mean(as.numeric(df_results$delta[8:9]))
dimResults[4] <- mean(as.numeric(df_results$delta[10:11]))
dimResults[5] <- mean(as.numeric(df_results$delta[12:13]))
df_dimResults$dim_results <- dimResults
```

```r
barplot(as.numeric(df_dimResults$dim_results), names.arg = as.list(df_dimResults$dim_names), cex.names = 0.69, legend.text = 'Gap (Smaller is better)')
```
